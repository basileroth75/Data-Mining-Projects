## Question 1
Calcul du coefficient Kappa de Cohen

```python
Pra=(625+500+150.)/1500
print'Pr(a) = ',Pra
Apos=(625+55+20.)/1500
print 'Apos = ',Apos
Bpos=(625+65+10.)/1500
print 'Bpos = ',Bpos
pos=Apos*Bpos
print 'pos =',pos
Aneg=(65+500+35.)/1500
Bneg=(55+500+40.)/1500
neg=Aneg*Bneg
print 'neg = ',neg
Aneu=(10.+40+150)/1500
Bneu=(20.+35+150)/1500
neu=Aneu*Bneu
print 'neu = ',neu
pre=neu+neg+pos
print 'pr(e) = ',pre
k=(Pra-pre)/(1-pre)
print 'k = ',k
```

Nous trouvons un coefficient Kappa de Cohen égale a 0,75.

###### B. Validation d'une distribution normale de la longueur de dent par un graphe Q-Q


```python
import numpy as np 
import pylab 
import scipy.stats as stats

vco5 = np.array([4.2,11.5,7.3,5.8,6.4,10,11.2,11.2,5.2,7])   
vc1 = np.array([16.5,16.5,15.2,17.3,22.5,17.3,13.6,14.5,18.8,15.5])
vc2=np.array([23.6,18.5,33.9,25.5,26.4,32.5,26.7,21.5,23.3,29.5])
ojo5=np.array([15.2,21.5,17.6,9.7,14.5,10,8.2,9.4,16.5,9.7])
oj1=np.array([19.7,23.3,23.6,26.4,20,25.2,25.8,21.2,14.5,27.3])
oj2=np.array([25.5,26.4,22.4,24.5,24.8,30.9,26.4,27.3,29.4,23])
stats.probplot(vco5, dist="norm", plot=pylab)
stats.probplot(vc1, dist="norm", plot=pylab)
stats.probplot(vc2, dist="norm", plot=pylab)
stats.probplot(ojo5, dist="norm", plot=pylab)
stats.probplot(oj1, dist="norm", plot=pylab)
stats.probplot(oj2, dist="norm", plot=pylab)
pylab.show()
```

D'Après le graphiques précédent, nous voyons que la correlation des courbe Q-Q se rapproche de 1, on peut donc conclure que la longueur de dents suit une distribution normal et cette distribution est différente selon le mode d'admisnistration de la vitamine.
