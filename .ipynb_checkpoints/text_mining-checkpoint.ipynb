```python
import glob
import string 
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 

ps = PorterStemmer() 
list_sentence = []
list_list_sentence = []
file_names = glob.glob("./Federalist/*.txt")
for file_name in file_names:
    print(file_name)
    with open(file_name, 'r') as file:
        data = file.read().replace('\n', '').lower()
        # print(data)
        remove_punctuation = data.translate(str.maketrans(dict.fromkeys(string.punctuation)))
        # print(remove_punctuation)
        remove_digits = ''.join([i for i in remove_punctuation if not i.isdigit()]) 
        # print(remove_digits)
        
        stop_words = set(stopwords.words('english')) 
        word_tokens = word_tokenize(remove_digits) 
        filtered_sentence = [w for w in word_tokens if not w in stop_words] 
        filtered_sentence = [] 
        for w in word_tokens: 
            if w not in stop_words: 
                filtered_sentence.append(ps.stem(w)) 
                
        # print(filtered_sentence)
        sentence = ' '.join(filtered_sentence)
        list_sentence.append(sentence)
        list_list_sentence.append(filtered_sentence)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

corpus = list_sentence
df = pd.DataFrame(data=corpus, columns=['sentences'])

vectorizer = CountVectorizer(vocabulary=['he', 'is', 'a', 'good', 'person', 'bad', 'student', 'hardworking'], min_df=0,token_pattern=r"(?u)\b\w+\b")
X = vectorizer.fit_transform(df['sentences'].values)
result = pd.DataFrame(data=X.toarray(), columns=vectorizer.get_feature_names())
print(result)

```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
 

df1 = pd.DataFrame({'Federalist18_Collaboration': [list_sentence[0]], 
                    'Federalist03_Jay': [list_sentence[1]], 
                    'Federalist17_Hamilton': [list_sentence[2]], 
                    'Federalist14_Madison': [list_sentence[3]], 
                    'Federalist05_Jay': [list_sentence[4]], 
                    'Federalist04_Jay': [list_sentence[5]]
                   })
 
# Initialize
vectorizer = TfidfVectorizer()
doc_vec = vectorizer.fit_transform(df1.iloc[0])
 
# Create dataFrame
df2 = pd.DataFrame(doc_vec.toarray().transpose(),index=vectorizer.get_feature_names())
 
# Change column headers
df2.columns = df1.columns
display(df2)
```
