# Use a Q-Q plot to check if an attribute follows a normal distribution

```python
import pandas as pd 

df_cochons = pd.read_csv("cochons.csv")
display(df_cochons.head(5))
```

```python
import numpy as np 
import statsmodels.api as sm 
import pylab as py 

sm.qqplot(df_cochons["len"]) 
py.show() 
```
